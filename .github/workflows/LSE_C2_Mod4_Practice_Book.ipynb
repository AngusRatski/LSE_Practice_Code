### Matplotlib plotting basics

```python
# Import numpy, pandas and matplotlib
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
# Create an empty figure
fig, ax = plt.subplots()
```

```python
# Assign values to axes with one positional argument
ax.plot([0, 10, 20, 30, 40])

# Inspect the new figure
fig
```

```python
ax.clear()# Clear the previous figure
ax.plot([0, 1, 2, 3, 4], [4, 3, 2, 1, 0])# Create figure with two positional arguments

# Inspect the new figure
fig
```

```python
# Specify three arguments = x-values, y-values, and marker.
ax.clear()
ax.plot([0, 1, 2, 3, 4], [4, 3, 2, 1, 0], 'og-')

# View figure.
fig
```

```python
# Specify various named arguments.
ax.clear()
ax.plot([0, 1, 2, 3, 4],
        [4, 3, 2, 1, 0],
        color='#94CEE4',
        marker='x',
        markeredgecolor='red',
        drawstyle='steps',
        linestyle='-')

# View figure.
fig
```

```python
# Create a plot using seaborn
sns.relplot(x=[1, 2, 3, 4], y=[4, 3, 2, 1], kind='line')
```

### Create plots using the penguins data set

```python
# Import the data into a dataframe
penguins = pd.read_csv('penguins.csv')

#Inspect the data and metadata
print(penguins.info())
print(penguins.describe())
penguins.head()
```

```python
sns.countplot(x='island', data=penguins, hue='species')
```

```python
# Create a simple pairplot.
sns.pairplot(penguins, hue='species')
```

```python
# Remove any redundant plots.
sns.pairplot(penguins,
             x_vars=['bill_length_mm', 'bill_depth_mm', 'flipper_length_mm'],
             y_vars=['bill_length_mm', 'bill_depth_mm'])
```

```python
# Create a boxplot of the weight distributions
order_species = penguins.groupby('species')['body_mass_g'].mean().iloc[::-1].index

sns.boxplot(x='species',
            y='body_mass_g',
            data=penguins,
            hue='sex',
            order=order_species,
            showfliers=False)
```

### Create a plot using the mpg data set

```python
# Import the mpg data into a dataframe
mpg = pd.read_csv('mpg.csv')

# Inspect the data and metadata
print(mpg.info())
mpg.head()
```

```python
# Create a lineplot
sns.lineplot(x='model_year', y='mpg', data=mpg)
```

A lineplot was created, which shows the change in miles per gallon (mpg) of car models over time. The shaded area around the line shows the confidence interval.

```python
# Create a barplot
sns.barplot(x='model_year', y='mpg', data=mpg)
```

```python
# Create a pointplot
sns.pointplot(x='model_year', y='mpg', hue='origin', data=mpg)
```

### Create barplots using the fitness data set

```python
# Import the data
fitness = pd.read_csv('daily_activity.csv')
```

```python
# Create a DataFrame with specified columns.
distance = fitness[['Id', 'ActivityDate', 'VeryActiveDistance',
                   'ModeratelyActiveDistance', 'LightActiveDistance',
                   'SedentaryActiveDistance']]

# View the DataFrame.
print(distance.head())


# Reshape the DataFrame from wide to long. 
distance_long = distance.melt(id_vars=['Id', 'ActivityDate'], 
                              var_name='DistanceType', value_name='Distance')

# View the DataFrame.
print(distance_long.head())


# Format values in the columns.
distance_long['DistanceType'] = distance_long['DistanceType'].str.replace('ActiveDistance', "")


# Add a filter to identify logged-in user.
individual = distance_long[distance_long['Id'] == 1503960366]


# Create a barplot.
sns.barplot(x='DistanceType', y='Distance', estimator=sum, data=individual)
```

```python
# Stacked barplots with Pandas.
individual.groupby('DistanceType')[['Distance']].sum().T.plot.bar(stacked=True)
```

```python
# Pair the data of two individuals.
pair = distance_long[distance_long['Id'].isin([8877689391, 1503960366])]

# Create a grouped barplot.
sns.barplot(x='DistanceType', y='Distance', hue='Id', estimator=sum, data=pair)
```
